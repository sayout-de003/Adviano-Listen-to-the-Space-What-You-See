```python
from google.colab import drive
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

def load_data(file_path):
    drive.mount('/content/drive')
    data = pd.read_csv(file_path)
    return data

# Example usage:
file_path = '/content/drive/My Drive/pulsar_data_test.csv'
data = load_data(file_path)
data = data.sample(frac=0.3, random_state=42)
```

```python
import pandas as pd
import numpy as np
from pydub import AudioSegment
from pydub.generators import Sine
```

```python
# Selecting features for sonification
# Use the exact column names with leading spaces
freq_feature = data[' Mean of the integrated profile']
vol_feature = data[' Standard deviation of the integrated profile']
```

```python
# Normalize the data for sonification
def normalize_data(series, min_val, max_val):
    norm = (series - series.min()) / (series.max() - series.min())
    return norm * (max_val - min_val) + min_val
```

```python

# Parameters for sonification
min_freq = 200  # Minimum frequency for the sound (Hz)
max_freq = 2000  # Maximum frequency for the sound (Hz)
min_vol = -30  # Minimum volume (dB)
max_vol = 0  # Maximum volume (dB)
duration_ms = 500  # Duration of each tone (ms)
```

```python
# Normalize features to frequency and volume ranges
data['Frequency (Hz)'] = normalize_data(freq_feature, min_freq, max_freq)
data['Volume (dB)'] = normalize_data(vol_feature, min_vol, max_vol)
```

```python
# Generate the sound
audio = AudioSegment.silent(duration=0)
```

```python

for index, row in data.iterrows():
    tone = Sine(row['Frequency (Hz)']).to_audio_segment(duration=duration_ms).apply_gain(row['Volume (dB)'])
    audio += tone
```

```python
# Export the audio
output_file = '/content/drive/My Drive/pulsar_sonification_new.wav'
audio.export(output_file, format='wav')
```
